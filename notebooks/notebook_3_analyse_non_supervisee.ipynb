### Visualisation PCA sur les données labellisées
La PCA est une technique qui permet de réduire le nombre de variables dans un jeu de données tout en conservant l’essentiel de l’information.
L’objectif de la visualisation PCA est d’évaluer la séparation réelle entre les classes connues. On veut voir si “Normal” et “Cancer” forment des groupes naturels dans l’espace des 512 features.

Fonctionnement :

* On regarde les directions où les données varient le plus.
* Imagine un nuage de points : la première composante est la direction dans laquelle les points s’étalent le plus.
* On crée de nouveaux axes (composantes) :
    * 1ère composante : capture le maximum d’information (variance).
    * 2ème composante : capture ce qu’il reste, dans une direction perpendiculaire.
* On projette les données sur ces nouveaux axes.
* Cela revient à représenter les données dans un espace “compressé”.

#### On enlève les labels de notre jeu labellisé

```python
X_labeled = df_labeled.drop(columns = "label")
y_labeled = df_labeled["label"]
```

#### On commence par normaliser nos données afin que chaque caractéristique puisse avoir une variance égale à 1.

```python
scaler = StandardScaler() # notre fonction de normalisation
scaler.fit(X_labeled) # nous entraînons le scaler sur le jeu de données
X_scaled_labeled = scaler.transform(X_labeled) # application de la rotation et la réduction de la dimension avec transform
```

#### Application du PCA pour réduire la dimensionnalité

```python
pca = PCA(n_components=2) # nombre de composantes principales que nous voulons afficher
pca.fit(X_scaled_labeled) # on fait tourner notre PCA

X_pca = pca.transform(X_scaled_labeled)
print("Original shape: {}".format(str(X_scaled_labeled.shape)))
print("Reduced shape : {}".format(str(X_pca.shape)))
```

* On note bien la réduction de dimension

#### Traçons les deux composantes principales 

```python
# points "normal"
plt.scatter(
    X_pca[y_labeled == 0, 0], 
    X_pca[y_labeled == 0, 1],
    color="blue", 
    label="Normal (0)",
    alpha=0.8
)

# points "cancer"
plt.scatter(
    X_pca[y_labeled == 1, 0], 
    X_pca[y_labeled == 1, 1],
    color="red", 
    label="Cancer (1)",
    alpha=0.8
)

plt.title("PCA - données labellisées uniquement")
plt.xlabel("Premier composant principal")
plt.ylabel("Deuxième composant principal")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

* On peut constater une bonne séparation naturelle de notre jeu. Les embeddings ResNet ont réussi à capturer des différences réelles entre les deux classes. On peut observer une concentration des éléments "normaux" sur la gauche et de "cancer" sur la droit avec une plus grande dispersion. Ce qui peut être expliqué par l'hétérogénéité des images cancéreuses, les variations entre les images sont plus grandes.

## Méthode DBScan

- L'algorithme fonctionne en définissant les grappes comme des régions denses séparées par des régions de moindre densité. Cette approche permet à DBSCAN de découvrir des grappes de forme arbitraire et d'identifier les valeurs aberrantes comme du bruit.

- DBSCAN s'articule autour de trois concepts clés :

* Points essentiels: Il s'agit de points qui ont au moins un nombre minimum d'autres points (MinPts) à une distance spécifiée (ε ou epsilon).
* Points frontières: Il s'agit de points qui se trouvent à une distance ε d'un point central mais qui n'ont pas de MinPts voisins eux-mêmes.
* Points de bruit: Il s'agit de points qui ne sont ni des points centraux ni des points frontières. Ils ne sont pas assez proches d'un groupe pour être inclus.

#### On récupère notre DataFrame

```python
df_clustering = pd.concat([
    df_labeled.drop(columns='label'),
    df_unlabeled],
    ignore_index=True)
```

#### On normalise nos données

```python
scaler = StandardScaler() # notre fonction de normalisation
scaler.fit(df_clustering) # nous entraînons le scaler sur le jeu de données
X_scaled_cluster = scaler.transform(df_clustering) # application de la rotation et la réduction de la dimension avec transform
```

#### On réduit les dimensions avec PCA

```python
pca = PCA(n_components=50) # nombre de composantes principales que nous voulons afficher
pca.fit(X_scaled_cluster) # on fait tourner notre PCA

X_pca_cluster = pca.transform(X_scaled_cluster)
print("Original shape: {}".format(str(X_scaled_cluster.shape)))
print("Reduced shape : {}".format(str(X_pca_cluster.shape)))
```

#### On calcul notre point epsilon en fonction du pgraphique k distance

```python
# Fonction du grpahique k distance
def plot_k_distance_graph(X, k):
    neigh = NearestNeighbors(n_neighbors=k)
    neigh.fit(X)
    distances, _ = neigh.kneighbors(X)
    distances = np.sort(distances[:, k-1])
    plt.figure(figsize=(10, 6))
    plt.plot(distances)
    plt.xlabel('Points')
    plt.ylabel(f'{k}-th nearest neighbor distance')
    plt.title('K-distance Graph')
    plt.show()
# Graph
plot_k_distance_graph(X_pca_cluster, k=5)
```

```python
dbscan = DBSCAN(eps=20, min_samples=5)
clusters = dbscan.fit_predict(X_pca_cluster)
print("Cluster :\n {}".format(clusters))
```

```python
# Visualize the results
plt.scatter(X_pca_cluster[:,0], X_pca_cluster[:,1], c=clusters, cmap="viridis")
plt.colorbar()
plt.show()
```

#### Avec cette méthode, la séparation ne se fait pas bien, on ne différencie pas les 2 classes.

## Calcul du score ARI

```python
X_labeled = df_labeled.drop(columns = "label")
y_labeled = df_labeled["label"]

scaler = StandardScaler()
X_labeled_scaled = scaler.fit_transform(X_labeled)

pca = PCA(n_components=50) # nombre de composantes principales que nous voulons afficher
pca.fit(X_labeled_scaled) # on fait tourner notre PCA

X_pca_cluster = pca.transform(X_labeled_scaled)

# Clustering KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_pca_cluster)

# Clustering DBSCAN
dbscan = DBSCAN(eps=20, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_pca_cluster)

# ARI
ari_kmeans = adjusted_rand_score(y_labeled, kmeans_labels)
ari_dbscan = adjusted_rand_score(y_labeled, dbscan_labels)

print("ARI KMeans :", ari_kmeans)
print("ARI DBSCAN :", ari_dbscan)
```
